## What is Genie?

**Genie** is Databricks' AI-powered analytics assistant that understands natural language questions.

**Key features:**
* **Natural language** - Ask questions in plain English
* **AI-generated SQL** - Automatically writes queries
* **Auto-visualization** - Creates appropriate charts
* **Knowledge store** - Curated semantic definitions improve accuracy
* **Semantic understanding** - Knows your data context via metadata and instructions
* **Conversational** - Follow-up questions

**Genie vs Traditional Dashboards:**

| Feature | Dashboards | Genie Spaces |
|---------|------------|---------------|
| **Query method** | Write SQL | Ask questions |
| **Visualization** | Manual config | AI-generated |
| **Best for** | Structured reports | Ad-hoc exploration |
| **Users** | SQL knowledge needed | No SQL needed |
| **Flexibility** | Fixed layout | Conversational |

### Step 12: View the Generated SQL Query

**Follow these steps:**

1. In Genie's response, click **"Show generated code"** to reveal the SQL
2. The actual SQL query appears

**Example SQL Genie might generate:**
```sql
SELECT 
  p.property_type,
  SUM(b.total_amount) AS total_revenue
FROM samples.wanderbricks.bookings b
JOIN samples.wanderbricks.properties p 
  ON b.property_id = p.property_id
WHERE b.status = 'confirmed'
GROUP BY p.property_type
ORDER BY total_revenue DESC
```

**What to notice:**
* Genie automatically joined bookings and properties
* Used SUM for "total revenue"
* Grouped by property_type
* Filtered to confirmed bookings
* Ordered by revenue (descending)

**SQL code features:**

* **Syntax highlighting** - Easy to read

## Part 4: Benchmarks & Evaluation

**Benchmarks** are question-answer pairs used to **evaluate** Genie's accuracy.

**What are benchmarks?**
* Curated question-answer pairs
* Include ground truth SQL ("SQL Answer")
* Used to **test and measure** Genie accuracy
* Track improvements over time as you refine your space

**Important:** Benchmarks are meant to **evaluate, not inform**, your Genie space. Genie does **not** use benchmark questions or example SQL from benchmarks to improve its context. To teach Genie, use **example SQL queries** and **instructions** instead.

**Why use benchmarks?**
* Measure Genie accuracy with a scored evaluation
* Test multiple phrasings of the same question
* Identify areas needing better instructions or SQL expressions
* Track improvement over time as you refine your knowledge store

---

### Step 15: Add a Benchmark Manually

**Follow these steps:**

1. Near the top of the Genie space, click **"Benchmarks"**
2. Click to add a new benchmark question
3. A benchmark creation form appears

**Create a new benchmark:**

4. **Enter the question:**
```
What was the average review rating by property type for each month?
```

5. **Enter the SQL Answer (ground truth):**
```sql
SELECT 
  YEAR(r.created_at) AS year,
  MONTH(r.created_at) AS month,
  p.property_type,
  AVG(r.rating) AS avg_rating
FROM samples.wanderbricks.reviews r
JOIN samples.wanderbricks.properties p 
  ON r.property_id = p.property_id
WHERE r.is_deleted = false
GROUP BY YEAR(r.created_at), MONTH(r.created_at), p.property_type
ORDER BY year, month, property_type
```

6. Click **"Save"**

**Benchmark best practices:**
* Cover common question patterns
* Include different complexity levels
* Test various tables and joins
* Add 2-4 **different phrasings** of the same question with the same SQL Answer
* Each benchmark question must have exactly one SQL Answer
* Questions without a SQL Answer require manual review

**Remember:** Benchmarks are for **evaluation only**. To teach Genie how to answer questions, use example SQL queries and instructions instead.

**Checkpoint:** Benchmark added manually

---

**Tip:** Most spaces should include between 2-4 phrasings of the same question to fully assess accuracy

### Step 16: Run Benchmark Evaluations

**Follow these steps:**

1. Near the top of the Genie space, click **"Benchmarks"**
2. You'll see your list of benchmark questions
3. Click **"Run benchmarks"** to test all questions
4. Or select specific questions and click **"Run selected"**

**What happens:**
* Genie processes each benchmark question
* Generates SQL for each
* Compares generated results to the SQL Answer you provided
* Scores each for accuracy

**Evaluation ratings:**

| Condition | Rating |
|-----------|--------|
| Generated SQL exactly matches your SQL Answer | ✅ **Good** |
| Result set exactly matches (even if SQL differs) | ✅ **Good** |
| Same data but sorted differently | ✅ **Good** |
| Numeric values round to same 4 significant digits | ✅ **Good** |
| Generated SQL produces empty result set or error | ❌ **Bad** |
| Result set includes extra columns | ❌ **Bad** |
| Single cell result is different | ❌ **Bad** |
| No SQL Answer provided or no exact match found | 👁️ **Manual review needed** |

**Review results:**

5. Click on any result to see:
   * Expected SQL (your SQL Answer)
   * Generated SQL (what Genie produced)
   * Differences highlighted

6. **Fix failures by improving your space:**
   * Update table descriptions
   * Add synonyms
   * Add or refine instructions and SQL expressions
   * Add example SQL queries

**Best practice:** Add multiple phrasings of the same question (2-4 phrasings) with the same SQL Answer to test accuracy across different wordings.

**Try this:**
1. Run evaluation on your benchmarks
2. Review the accuracy ratings
3. Click on any failed benchmarks
4. Understand why they failed

### **Challenge 2: Create Multiple Benchmarks**

**Your task:**

Create 3 more benchmarks covering different question types.

**Benchmark 1: Simple aggregation**
* **Question:** "How many confirmed bookings do we have?"
* **SQL Answer:**
```sql
SELECT COUNT(*) AS booking_count
FROM samples.wanderbricks.bookings
WHERE status = 'confirmed'
```

**Benchmark 2: Time-series analysis**
* **Question:** "Show me daily booking revenue for the last 30 days"
* **SQL Answer:**
```sql
SELECT 
  DATE(created_at) AS booking_date,
  SUM(total_amount) AS daily_revenue
FROM samples.wanderbricks.bookings
WHERE status = 'confirmed'
  AND created_at >= CURRENT_DATE() - INTERVAL 30 DAYS
GROUP BY DATE(created_at)
ORDER BY booking_date
```

**Benchmark 3: Multi-table join**
* **Question:** "Which properties have the highest average review ratings?"
* **SQL Answer:**
```sql
SELECT 
  p.title AS property_name,
  p.property_type,
  AVG(r.rating) AS avg_rating,
  COUNT(r.review_id) AS review_count
FROM samples.wanderbricks.properties p
JOIN samples.wanderbricks.reviews r 
  ON p.property_id = r.property_id
WHERE r.is_deleted = false
GROUP BY p.title, p.property_type
HAVING COUNT(r.review_id) >= 5
ORDER BY avg_rating DESC
LIMIT 10
```

**Requirements:**
1. Add all 3 benchmarks (Benchmarks tab → Add)
2. Add 2-4 alternative phrasings for each question
3. Run evaluation via **"Run benchmarks"**
4. Review results (Good / Bad / Manual review needed)

**Remember:** These benchmarks evaluate Genie's accuracy but do not teach it. To improve accuracy, add example SQL queries and SQL expressions.

## Part 5: Custom Instructions 📝

**Instructions** teach Genie about your business logic and data semantics.

**What are instructions?**
* Custom rules and context for Genie
* Accessed via **Configure → Instructions**
* Three types, in order of preference:

**1. SQL expressions** (preferred for business terms):
* Define KPIs, metrics, filters, and dimensions
* See Part 7 for details

**2. Example SQL queries** (preferred for complex questions):
* Show Genie how to approach common question formats
* See Part 8 for details

**3. Text instructions** (use as a last resort):
* Plain text guidance for Genie
* Use only when SQL expressions and examples cannot address the need
* Accessed via **Configure → Instructions → Text** tab
* The entire text block counts as one instruction

**Instruction limit:** Up to **100 instructions** total per Genie space.

**Use cases for text instructions:**
* Natural language guidance (e.g., "When users ask about customer performance without specifying a time range, ask them to clarify")
* Rounding rules (e.g., "Always round percentages to two decimal places")
* Avoid using text instructions for metrics or filters that can be expressed through SQL

## Part 7: SQL Expressions 📊

**SQL expressions** are reusable business logic definitions in the **knowledge store** that Genie can reference.

**What are SQL expressions?**
* Structured definitions for business terms and KPIs
* Reusable across questions
* Encode business logic
* Three types: **Measures**, **Filters**, and **Dimensions**

**SQL expression types:**
* 📊 **Measures** - KPIs and metrics (e.g., total revenue, conversion rate). Aggregation calculations.
* 🔍 **Filters** - Common filtering conditions (e.g., "high value orders"). Boolean conditions.
* 📏 **Dimensions** - Attributes for grouping/analyzing (e.g., order year, customer segment). Row-level transformations.

**Use cases:**
* Business definitions ("high value", "active customer")
* Complex filters
* Calculated metrics
* Data quality rules

**Best practice:** Use SQL expressions for common business terms. Use example SQL queries for complex multi-step questions. Use text instructions only as a last resort.

## 🔧 Troubleshooting Genie Issues

**Genie gives wrong answers:**
* Check table descriptions are accurate
* Add synonyms for columns
* Add SQL expressions for business logic
* Add example SQL queries for similar questions
* Review generated SQL via "Show generated code"
* Edit SQL and click "Add as instruction" to teach Genie

**Genie can't find the right table:**
* Improve table descriptions
* Add column synonyms
* Mention use cases in description
* Ensure table is added to space
* Hide confusing or irrelevant columns

**Joins are incorrect:**
* Define foreign key references in Unity Catalog when possible
* Configure explicit join relationships in the knowledge store
* Verify join conditions are correct
* Provide example SQL queries that demonstrate correct joins
* Consider pre-joining into a view as a last resort

**Genie doesn't understand terminology:**
* Add SQL expressions for business terms (measures, filters, dimensions)
* Add example SQL queries showing correct usage
* Add synonyms to columns
* Use text instructions only if SQL-based options are insufficient

**Low benchmark accuracy:**
* Review failed benchmarks (Bad / Manual review needed)
* Update SQL Answer if needed
* Add example SQL queries for failure patterns
* Add SQL expressions for misunderstood terms
* Improve table/column configuration
* Review knowledge mining recommendations

**Slow responses:**
* Check SQL warehouse is running
* Optimize dataset queries
* Reduce data volume with filters
* Use appropriate warehouse size
