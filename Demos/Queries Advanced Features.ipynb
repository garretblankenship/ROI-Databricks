## 1. Query Parameters (Widgets) üéØ

**Query parameters** make your queries dynamic and interactive by allowing users to provide input values.

**What are parameters?**
* Input fields that appear above your query
* Users can change values without editing SQL
* Perfect for filters, date ranges, thresholds
* Also called "widgets" or "filters"

**Use cases:**
* Date range selection
* Location filters
* Threshold values
* Category selection
* Dashboard interactivity

---

### üéØ Step 1: Create a Text Parameter

**Try this:**

1. In SQL Editor, write this query:
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE pickup_zip = {{zip_code}}
GROUP BY pickup_zip
```

4. After typing `{{zip_code}}`, a **parameter widget** appears above the query

5. The widget shows:

   * Input field (empty by default)
   * Gear icon for settings

**Try it now:**

### üìä Step 3: Different Parameter Types

**Databricks Queries supports multiple parameter types:**

**1. Text**
* Free-form text input
* Use for: IDs, names, search terms
* Example: `{{customer_name}}`

**2. Number**
* Numeric input only
* Use for: Thresholds, limits, amounts
* Example: `{{min_fare}}`

**3. Dropdown**
* Select from predefined options

* Start and end date pickers

* Example: `{{date_range.start}}` and `{{date_range.end}}`

**6. Query-based Dropdown**
* Options populated from another query
* Use for: Dynamic lists
* Example: List of active customers

### ‚úÖ Query Parameters - Best Practices

**Naming conventions:**
* ‚úÖ Use descriptive names: `start_date`, not `d1`
* ‚úÖ Use snake_case: `min_fare`, not `minFare`
* ‚úÖ Be consistent across queries
* ‚úÖ Match business terminology

**Setting defaults:**
* ‚úÖ Always provide default values
* ‚úÖ Use sensible defaults (last 30 days, common filters)
* ‚úÖ Defaults should return meaningful results
* ‚úÖ Test with default values

**Choosing types:**
* ‚úÖ **Text** - When any value is valid
* ‚úÖ **Number** - For numeric thresholds
* ‚úÖ **Dropdown** - When options are limited
* ‚úÖ **Date** - For date filters
* ‚úÖ **Date Range** - For time periods

**User experience:**
* ‚úÖ Use clear titles ("Start Date", not "start_date")
* ‚úÖ Limit number of parameters (5-7 max)
* ‚úÖ Group related parameters
* ‚úÖ Provide helpful defaults

**Common patterns:**
```sql
-- Date range filter
WHERE date >= '{{start_date}}' AND date <= '{{end_date}}'

-- Threshold filter
WHERE amount >= {{min_amount}}

-- Category filter
WHERE status = '{{status}}'

-- Dynamic LIMIT
LIMIT {{row_limit}}
```

## 2. Multiple Statement Results üìä

**Multiple statement queries** let you run several SQL statements in one query and see all results.

**What is it?**
* Write multiple SQL statements separated by semicolons
* Run them all at once
* Each statement produces its own result set
* Results appear in separate tabs

**Use cases:**
* Data quality checks (multiple validation queries)
* Before/after comparisons
* Multi-step analysis
* Comprehensive reports

---

### üìù Step 7: Write Multiple SQL Statements

**Try this:**

1. Write multiple statements separated by semicolons:

```sql
-- Statement 1: Overall statistics
SELECT 
  COUNT(*) AS total_trips,
  AVG(fare_amount) AS avg_fare,
  AVG(trip_distance) AS avg_distance
FROM samples.nyctaxi.trips
LIMIT 1;

-- Statement 2: Top pickup locations
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10;

-- Statement 3: Fare distribution
SELECT 
  CASE 
    WHEN fare_amount < 10 THEN 'Under $10'
    WHEN fare_amount < 20 THEN '$10-$20'
    WHEN fare_amount < 50 THEN '$20-$50'
    ELSE '$50+'
  END AS fare_range,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
GROUP BY fare_range
ORDER BY fare_range;
```

2. Notice the **semicolons** (`;`) separating statements

3. Each statement is independent

**Try it now:**
1. Copy the query above into SQL Editor
2. Notice you have 3 separate SELECT statements

---

**üí° Tip:** Add comments to explain what each statement does

### üéØ Step 9: Combine Multiple Statements with Parameters

**Try this:**

Use parameters across multiple statements:

```sql
-- Statement 1: Summary for date range
SELECT 
  'Summary' AS report_section,
  COUNT(*) AS total_trips,
  SUM(fare_amount) AS total_revenue,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= '{{start_date}}'
  AND DATE(tpep_pickup_datetime) <= '{{end_date}}';

-- Statement 2: Daily breakdown
SELECT 
  DATE(tpep_pickup_datetime) AS trip_date,
  COUNT(*) AS trip_count,
  SUM(fare_amount) AS daily_revenue
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= '{{start_date}}'
  AND DATE(tpep_pickup_datetime) <= '{{end_date}}'
GROUP BY DATE(tpep_pickup_datetime)
ORDER BY trip_date;

-- Statement 3: Top locations in period
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= '{{start_date}}'
  AND DATE(tpep_pickup_datetime) <= '{{end_date}}'
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10;
```

**What happens:**
* Same parameters (`start_date`, `end_date`) used in all statements
* Change parameters once, all statements update
* Creates a comprehensive report

## 4. Query Scheduling üîÑ

**Query scheduling** lets you run queries automatically on a schedule.

**What is scheduling?**
* Queries run automatically at specified times
* No manual execution needed
* Results are cached and available
* Perfect for reports and dashboards

**Use cases:**
* Daily reports (run every morning)
* Hourly data refreshes
* Weekly summaries
* Real-time dashboard updates
* Automated data quality checks

### üìÖ Step 16: Schedule Frequency Options

**Available frequencies:**

**1. Never (Manual only)**
* Query only runs when you click Run
* No automatic execution
* Default for new queries

**2. Minutes**
* Every N minutes (e.g., every 5, 15, 30 minutes)
* Use for: Real-time dashboards, monitoring
* Example: Every 15 minutes

**3. Hourly**
* Every N hours (e.g., every 1, 2, 6 hours)
* Use for: Frequent updates, operational dashboards
* Example: Every hour at :00

**4. Daily**
* Every day at specific time
* Use for: Daily reports, morning summaries
* Example: Every day at 8:00 AM

**5. Weekly**
* Specific days of week at specific time
* Use for: Weekly reports, summaries
* Example: Every Monday at 9:00 AM

**6. Monthly**
* Specific day of month at specific time
* Use for: Monthly reports, end-of-month summaries
* Example: 1st of month at 8:00 AM

**7. Custom (Cron)**
* Advanced cron expression
* Use for: Complex schedules
* Example: `0 8 * * 1-5` (weekdays at 8 AM)

**üí° Tip:** Choose frequency based on how often data updates and how fresh results need to be

### ‚ö†Ô∏è Step 20: Alert Condition Types

**Available conditions:**

**1. Greater than (>)**
* Triggers when value exceeds threshold
* Example: Error count > 100
* Use for: Upper limits, maximums

**2. Greater than or equal (>=)**
* Triggers when value meets or exceeds threshold
* Example: Revenue >= 10000
* Use for: Targets, goals

**3. Less than (<)**
* Triggers when value falls below threshold
* Example: Active users < 1000
* Use for: Lower limits, minimums

**4. Less than or equal (<=)**
* Triggers when value meets or falls below threshold
* Example: Inventory <= 10
* Use for: Stock alerts, capacity

**5. Equals (=)**
* Triggers when value exactly matches
* Example: Failed jobs = 0
* Use for: Specific states

**6. Not equals (!=)**
* Triggers when value doesn't match
* Example: Status != 'healthy'
* Use for: Error conditions

**Query requirements:**
* Must return a **single numeric value**
* Use aggregations: COUNT, SUM, AVG, MAX, MIN
* Return one row, one column

**Example queries for alerts:**
```sql
-- Count-based alert
SELECT COUNT(*) AS error_count FROM errors WHERE date = CURRENT_DATE()

-- Percentage-based alert
SELECT (failed_orders / total_orders * 100) AS failure_rate FROM metrics

-- Threshold-based alert
SELECT AVG(response_time) AS avg_response FROM logs WHERE timestamp > NOW() - INTERVAL 1 HOUR
```

* Trigger: `agg`
* Code:
```sql
SELECT 
  ${1:group_column},
  COUNT(*) AS row_count,
  SUM(${2:amount_column}) AS total,
  AVG(${2:amount_column}) AS average
FROM ${3:table_name}
GROUP BY ${1:group_column}
ORDER BY row_count DESC
```

* Trigger: `nullcheck`
* Code:
```sql
SELECT 
  '${1:column_name}' AS column_name,
  COUNT(*) AS total_rows,
  SUM(CASE WHEN ${1:column_name} IS NULL THEN 1 ELSE 0 END) AS null_count,
  ROUND(SUM(CASE WHEN ${1:column_name} IS NULL THEN 1 ELSE 0 END) * 100.0 / COUNT(*), 2) AS null_percentage
FROM ${2:table_name}
```

* Trigger: `topn`
* Code:
```sql
SELECT 
  ${1:column},
  COUNT(*) AS count
FROM ${2:table}
GROUP BY ${1:column}
ORDER BY count DESC
LIMIT ${3:10}
```
