# Uploading Data via the Workspace UI

## Instructions

## Overview

This hands-on demo walks you through creating a Unity Catalog namespace and uploading data files using the Databricks UI. You'll create a catalog and schema, then use the built-in upload wizard to import CSV files into managed tables — learning how to configure column types and handle multi-line data along the way.

**What you'll learn:**
* Creating catalogs and schemas via the Catalog Explorer
* Uploading CSV files through the UI
* Configuring column data types during upload
* Handling multi-line CSV rows
* Reviewing table metadata in the Catalog UI

1. In the Left Sidebar, click on `Catalog`.
![alt text](./Assets/21/01-catalog-sidebar.png)

2. In the Catalog Explorer, click the **Create Catalog** button.
![alt text](./Assets/21/02-create-catalog-btn.png)

3. In the **Create a new catalog** dialog, configure the following:
   - **Name:** `ecommerce`
   - **Type:** `Standard`
   - **Storage Location:** Leave as default (Metastore Default) or select your preferred managed location
   - **Comment:** *(optional)* Add a description for the catalog
4. Click **Create** to finish.
![alt text](./Assets/21/03-catalog-config.png)

5. With the `ecommerce` catalog selected in the Catalog Explorer, click **Create Schema**.
6. Enter `demo` as the schema name and click **Create**.
![alt text](./Assets/21/04-create-schema.png)

7. Navigate into `ecommerce` → `demo` in the Catalog Explorer.
8. Click the **Create Table** button from within the schema view.
   > **Note:** We're using the Create Table option inside the Catalog Explorer rather than the top-level `New` menu. This approach pre-fills the catalog and schema for you.

![alt text](./Assets/21/05-create-table-catalog.png)

## Uploading sales.csv

9. In the upload wizard, drag and drop (or browse for) the `sales.csv` file from the `Assets/21` folder.
10. Review the auto-detected columns and data types in the preview.
![alt text](./Assets/21/06-upload-sales.png)

11. Find the `Product line` column in the column list.
12. Click on its data type and change it from `STRING` to **category**.
13. Review any other columns as needed, then click **Create Table**.
![alt text](./Assets/21/07-product-line-category.png)

## Uploading reviews.csv

14. Navigate back to `ecommerce` → `demo` and click **Create Table** again.
15. Upload the `reviews.csv` file.
16. Notice that some rows may not parse correctly — review text fields contain line breaks that span multiple lines.
17. In the upload options, enable **Rows can span multiple lines** to handle multi-line field values.
18. Review the corrected preview and click **Create Table**.
![alt text](./Assets/21/08-reviews-multiline.png)

## Uploading branch.csv

19. Navigate back to `ecommerce` → `demo` and click **Create Table** once more.
20. Upload the `branch.csv` file.
21. Review the auto-detected columns — this file should upload cleanly with default settings.
22. Click **Create Table**.
![alt text](./Assets/21/09-upload-branch.png)

## Upload Wizard: Unity Catalog Integration

23. Let's review the key Unity Catalog components of the upload wizard:
    - **Catalog** dropdown: Determines which catalog the table is created in (we used `ecommerce`)
    - **Schema** dropdown: Determines which schema within the catalog (we used `demo`)
    - **Table Name**: Auto-populated from the file name, but can be customized
    - **Column Types**: Can be modified before creating the table
    - **Advanced Options**: Includes settings for delimiters, headers, encoding, and multi-line handling

> **Key Takeaway:** By creating the table from within the Catalog Explorer (step 8), the catalog and schema fields were pre-filled. If you use the top-level `New` → `Create Table` menu instead, you'll need to manually select these.

![alt text](./Assets/21/10-wizard-uc.png)

## Reviewing Table Information in the Catalog UI

24. In the Catalog Explorer, navigate to one of your newly created tables (e.g., `ecommerce.demo.sales`).
25. Explore the available tabs:
    - **Overview**: Table description, tags, and general metadata
    - **Columns**: Column names, data types, and comments
    - **Sample Data**: Preview of the first rows of the table
    - **Details**: Storage information, table properties, creation date, and owner
    - **Permissions**: Access controls managed through Unity Catalog
    - **Lineage**: Data lineage showing upstream and downstream dependencies
    - **Quality**: Data quality rules and monitoring (if configured)

![alt text](./Assets/21/11-table-info.png)
