## What is Databricks SQL Editor?

The **Databricks SQL Editor** is a web-based IDE for writing and executing Databricks queries against your data lakehouse.

**Key features:**
* Schema Browser - Explore tables and columns
* Intelligent code editor - Autocomplete, syntax highlighting
* Tab management - Work on multiple queries
* Fast execution - Serverless SQL warehouses
* Query library - Save and share queries
* Results visualization - Tables and charts

---

## Getting Started

**To access SQL Editor:**
1. Click **SQL Editor** in the left sidebar (open in new tab)
2. Or navigate to `/sql/editor` in your workspace

**Prerequisites:**
* Access to a Databricks workspace
* SQL warehouse (serverless or classic)
* Permissions to query data

## 1. Schema Browser

The **Schema Browser** is your data catalog - it shows all available databases, tables, and columns.

**Location:** Left sidebar of SQL Editor

---

### Interface Overview

The Schema Browser has a hierarchical structure:

```
Catalog (e.g., main, samples)
  └─ Schema/Database (e.g., default, nyctaxi)
      └─ Table (e.g., customers, orders)
          └─ Columns (name, type, description)
```

**Three-level namespace:**
* **Catalog** - Top-level container (Unity Catalog)
* **Schema** - Database/schema within catalog
* **Table** - Actual data table

**Full table name:** `catalog.schema.table`

---

### Step 1: Explore Catalogs

**Try this:**

1. Look at the **Schema Browser** on the left side
2. You'll see a list of catalogs (e.g., `workspace`, `samples`)
3. Click the **dropdown arrow** next to a catalog name to expand it

**What you'll see:**
* **workspace** - Your workspace's default catalog
* **samples** - Sample datasets provided by Databricks
* Other catalogs - If your organization has created them

**Try it now:**
* Expand the `samples` catalog
* You should see schemas like `nyctaxi`, `tpch`

---

**Tip:** Catalogs organize data by domain or team (e.g., marketing, sales, engineering)

### Step 2: Explore Schemas

**Try this:**

1. Expand a catalog (e.g., `samples`)
2. You'll see a list of schemas (databases)
3. Click the **dropdown arrow** next to a schema to expand it

**Example: samples.nyctaxi**
* Contains New York City taxi trip data
* Multiple tables with different data

**Try it now:**
* Expand `samples.nyctaxi`
* You should see tables like `trips`

---

**Tip:** Schemas group related tables together (e.g., all sales tables in one schema)

### Step 3: Explore Tables

**Try this:**

1. Expand a schema (e.g., `samples.nyctaxi`)
2. You'll see a list of tables
3. Click on a **table name** to see its details

**What you'll see when you click a table:**
* **Columns** - Column names and data types
* **Sample data** - Preview of actual data

**Try it now:**
* Click on `samples.nyctaxi.trips`
* Explore the columns and sample data

---

**Tip:** Hover over a table name to see a quick preview tooltip

### Step 4: Quick Insert Table/Column Names

**Try this:**

You can quickly insert table and column names into your query:

**Method 1: Click to insert**
1. Click the double arrows next to a **table name** in Schema Browser
2. The full table name appears in your query!

**Method 2: Drag and drop**
1. Click and **drag** a table name from Schema Browser
2. **Drop** it into your query editor
3. The table name is inserted

**Method 3: Copy/Paste**
1. Click the 3 dots next to a **table name** in Schema Browser
2. Click `Copy Table Path`, then paste in the query editor
3. The table name is pasted

**Try it now:**
1. Clear your query editor
2. Type: `SELECT * FROM `
3. Use Schema Browser to insert `samples.nyctaxi.trips`
4. Your query should be: `SELECT * FROM samples.nyctaxi.trips`

---

**Tip:** This prevents typos in table names!

### Step 5: Search for Tables

**Try this:**

The Schema Browser has a **search box** at the top:

1. Look for the **search icon** at the top of Schema Browser
2. Click it to open the search box
3. Type a table name (e.g., "trips")
4. Results show matching tables across all catalogs

**Search features:**
* Searches across all catalogs and schemas
* Matches table names and descriptions
* Shows full path (catalog.schema.table)
* Click result to navigate to that table

**Try it now:**
1. Click the search icon in Schema Browser
2. Search for "trips"
3. Click on a result to see that table

---

**Tip:** Use search when you know the table name but not its location

## 2. Query Editor Features

The **Query Editor** is where you write your SQL code. It has intelligent features to help you write queries faster and with fewer errors.

**Location:** Center panel of SQL Editor

---

### Step 6: Autocomplete (IntelliSense)

The editor provides **intelligent autocomplete** as you type.

**Try this:**

1. Start typing in the query editor: `SELECT * FROM sam`
2. A dropdown menu appears with suggestions
3. You'll see:
   * Matching catalogs (samples)
   * Matching tables
   * SQL keywords
4. Use **arrow keys** to navigate suggestions
5. Press **Tab** or **Enter** to accept

**Autocomplete triggers:**
* After typing a few characters
* After typing `.` (dot) for catalog.schema.table
* After `FROM`, `JOIN`, `WHERE` keywords
* When typing column names

**Try it now:**
```sql
SELECT * FROM samples.nyc
```
* Type this and watch autocomplete suggest `nyctaxi`
* Press Tab to complete

### Step 7: Column Name Autocomplete

Autocomplete also works for **column names**!

**Try this:**

1. Type a query with a table:
```sql
SELECT 
  
FROM samples.nyctaxi.trips
```
2. Put your cursor after `SELECT`
3. Start typing a column name (e.g., `fare`)
4. Autocomplete suggests matching columns: `fare_amount`
5. Press Tab to insert

**Column autocomplete features:**
* Shows columns from tables in your query
* Shows data types next to column names
* Filters as you type
* Prevents typos

**Try it now:**
1. Write: `SELECT fare FROM samples.nyctaxi.trips`
2. Watch autocomplete suggest `fare_amount`
3. Complete the query and add `LIMIT 10`

---

**Tip:** Autocomplete knows which columns belong to which table in multi-table queries

   
### Step 8: Syntax Highlighting

The editor uses **color coding** to make SQL easier to read.

**Try this:**

Type this query and observe the colors:
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE fare_amount > 10
GROUP BY pickup_zip
LIMIT 10
```

**Benefits:**
* Easier to read
* Spot errors quickly
* Understand query structure
* Identify keywords vs identifiers

---

### Step 9: Format Your SQL Code

The editor can **automatically format** your SQL for better readability.

**Try this:**

1. Write a messy query (all on one line):
```sql
SELECT pickup_zip,COUNT(*) AS trip_count,AVG(fare_amount) AS avg_fare FROM samples.nyctaxi.trips WHERE fare_amount>10 GROUP BY pickup_zip LIMIT 10
```

2. Click on the 3 dots in the top right of the editor
3. Select `Edit -> Format Query`
4. Or use keyboard shortcut:
   * **Windows/Linux:** `Ctrl+Shift+F`
   * **Mac:** `Command+Shift+F`

**Result:**
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE fare_amount > 10
GROUP BY pickup_zip
LIMIT 10
```

**Try it now:**
1. Write a messy query
2. Format it using the keyboard shortcut
3. See how much more readable it becomes!

---

**Tip:** Format your queries regularly for better readability and collaboration

### Step 10: Error Detection and Warnings

The editor highlights **syntax errors** as you type.

**Try this:**

1. Type an invalid query:
```sql
SELECT * FORM samples.nyctaxi.trips
```
(Notice: "FORM" instead of "FROM")

2. You'll see:
   * **Red squiggly underline** under or near the error
   * **Hover** over it to see the error message
   * Error message explains what's wrong

**Common errors detected:**
* Misspelled keywords (FORM instead of FROM)
* Missing commas
* Unclosed quotes
* Invalid syntax
* Unknown table names (sometimes)

**Try it now:**
1. Type: `SELCT * FROM samples.nyctaxi.trips`
2. Hover over the red underline
3. Fix the error to `SELECT`

---

**Tip:** Fix errors before running - saves time and compute resources!

### Step 11: Adding Comments

Comments help document your queries.

**Single-line comments:**
```sql
-- This is a single-line comment
SELECT * FROM samples.nyctaxi.trips
```

**Multi-line comments:**
```sql
/*
  This is a multi-line comment
  You can write multiple lines
  Useful for longer explanations
*/
SELECT * FROM samples.nyctaxi.trips
```

**Try this:**

1. Add comments to your query:
```sql
-- Query: Analyze taxi trips by pickup location
-- Author: Your Name
-- Date: 2026-01-20

SELECT 
  pickup_zip,
  COUNT(*) AS trip_count  -- Count trips per location
FROM samples.nyctaxi.trips
WHERE fare_amount > 10  -- Filter expensive trips
GROUP BY pickup_zip
LIMIT 10
```

**Best practices:**
* Explain complex logic
* Document assumptions
* Add query metadata (author, date, purpose)
* Comment out code for testing

---

**Tip:** Use `Ctrl+/` (Windows/Linux) or `Cmd+/` (Mac) to quickly comment/uncomment lines

## 3. Tab Management

Work on **multiple queries simultaneously** using tabs - just like a web browser!

**Location:** Top of the query editor area

---

### Step 12: Create New Query Tabs

**Try this:**

1. Look at the top of the query editor
2. Find the **"+"** button next to existing tabs
3. Click it to create a new tab

**What happens:**
* A new untitled tab opens
* Empty query editor ready for your SQL
* Independent from other tabs

**Try it now:**
1. Create a new tab using the + button
2. You should now have 2 tabs open

---

**Tip:** Use separate tabs for different analyses or queries you're working on

### Step 13: Close Tabs

Databricks queries in the Databricks SQL Editor auto-save as you type. Your changes are persisted automatically, there's no manual save button needed. This applies to both the query content and any visualizations or parameters you've added.

**Try this:**

1. Hover over a tab
2. Click the **"X"** that appears
3. Tab closes

---

**Tip:** Close unused tabs to keep your workspace organized

### Step 14: Rename Query Tabs

**Try this:**

Give your queries descriptive names:

1. Click on the **tab name** (e.g., "Untitled Query")
2. The name becomes editable
3. Type a new name (e.g., "Daily Sales Analysis")
4. Press **Enter** to save

**Naming best practices:**
* Use descriptive names ("Customer Revenue Report")
* Include purpose ("Debug - Order Duplicates")
* Add date if relevant ("Q4 2024 Sales")
* Avoid generic names ("Query 1", "Test")

---

**Tip:** Good names help you find queries later in your query history

## 4. Running Queries & Viewing Results

Learn how to execute your queries and work with results.

---

### Step 15: Select a SQL Warehouse

Before running queries, you need to select a **SQL warehouse** (compute resource).

**Try this:**

1. Look at the **top-right** of the SQL Editor
2. Find the **warehouse dropdown** (shows current warehouse or "Select warehouse")
3. Click to dropdown
4. You'll see available warehouses:
   * **Serverless** - Fast, auto-scaling (recommended)
   * **Classic** - Traditional SQL warehouses
5. Select a warehouse

**Warehouse states:**
* **Running** - Ready to execute queries
* **Starting** - Warming up (wait a moment)
* **Stopped** - Click to start

**Try it now:**
1. Select a serverless warehouse (if available)
2. Wait for it to show "Running" status
3. You're ready to execute queries!

---

**Tip:** Serverless warehouses start instantly and scale automatically. You'll also need a SQL warehouse to be able to navigate the Catalog Browser

### Step 16: Execute Your Query

**Try this:**

Write and run a simple query:

1. Type this query:
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10
```

2. Click the **"Run"** button (blue button, top-left of editor)
3. Or use keyboard shortcut:
   * **Windows/Linux:** `Ctrl+Enter`
   * **Mac:** `Cmd+Enter`

**What happens:**
* Query is sent to the SQL warehouse
* Execution starts immediately
* Progress indicator shows status
* Results appear below the editor

**Try it now:**
1. Write the query above
2. Press `Ctrl+Enter` / `Cmd+Enter` to run
3. Wait for results to appear

---

**:** `Ctrl+Enter` / `Cmd+Enter` is the fastest way to run queries!

### Step 17: Run Selected SQL Only

You can run **part of your query** without running everything.

**Try this:**

1. Write multiple queries in one tab:
```sql
-- Query 1
SELECT COUNT(*) FROM samples.nyctaxi.trips;

-- Query 2
SELECT * FROM samples.nyctaxi.trips LIMIT 5;

-- Query 3
SELECT AVG(fare_amount) FROM samples.nyctaxi.trips;
```

2. **Highlight** Query 2 with your mouse
3. Click **"Run"** or press `Ctrl+Enter` / `Cmd+Enter`
4. Only the selected query runs!

**Use cases:**
* Test individual queries in a multi-query file
* Debug specific parts
* Run queries in different order
* Skip expensive queries

**Try it now:**
1. Write the 3 queries above
2. Highlight just the second query
3. Run only that query

---

**Tip:** This is great for testing and debugging complex query files

### Step 18: View Query Results

After running a query, results appear in the **Results panel** below the editor.

**Try this:**

1. Run a query (any query)
2. Look at the **Results panel** below
3. You'll see:
   * **Table view** - Data in rows and columns
   * **Row count** - Number of rows returned
   * **Execution time** - How long it took

**Result panel features:**
* **Scrollable** - Scroll horizontally and vertically
* **Sortable** - Click column headers to sort
* **Resizable** - Drag the divider to resize
* **Collapsible** - Minimize to see more of your query

**Try it now:**
1. Run: `SELECT * FROM samples.nyctaxi.trips LIMIT 100`
2. Explore the results:
   * Scroll through rows
   * Click column headers to sort
   * Resize the results panel

---

**Tip:** Results are limited to 1000 rows by default in the preview

Export your results for use in other tools.

**Try this:**

3. Click the **down arrow** next to the word `Table`
4. Download options:
   * **CSV** - Comma-separated values
   * **Excel** - .xlsx format (if available)

**Download options:**
* **First 1000 rows**
* **All rows**

**Try it now:**
1. Run: `SELECT * FROM samples.nyctaxi.trips LIMIT 50`
2. Click the down arrow next to the word Table

Learn how to save your work and keep your queries organized.

### Step 20: Clone (Duplicate Query)

Create a copy of an existing query.

**Try this:**

1. Open a saved query
2. Click the **three dots menu** (⋮) in the top right of the SQL Editor
3. Select `File -> Clone`
4. Enter a new name
5. Choose a folder
6. Click "Clone"

**Use cases:**
* Create variations of a query
* Make a backup before major changes
* Share a template with your team
* Create monthly/weekly versions

**Example:**
* Original: "Sales Report - January"
* Clone: "Sales Report - February"
* Modify dates in the new copy

**Try it now:**
1. Open your saved query
2. Use "Clone" to create "My First Query - Copy"
3. You now have two separate queries

### Step 22: Organize Queries in the Workspace Browser

Queries are **workspace objects** stored in your user home folder, just like notebooks and files. You organize them using the **Workspace browser**.

**Important:** The **Queries** sidebar (`Queries` in the left nav) is a **filtered list view** for finding queries by name, tag, or owner. It does **not** support creating or managing folders. All folder organization happens in the **Workspace browser**.

**What happens when you create a new query:**
* New queries created from the SQL Editor are **continuously autosaved**
* They initially appear in a **Drafts** section until explicitly saved to a workspace location
* Use the **Save** button to persist the query to a specific folder in your workspace

**Organize queries using the Workspace browser:**

1. Click **Workspace** in the left sidebar
2. Navigate to your **Home** folder (or any folder you have access to)
3. Create folders here to group your queries:
   * Right-click → **Create** → **Folder**
   * Name it (e.g., "Sales Analysis", "Data Quality Checks")
4. Move queries into folders:
   * Right-click a query → **Move**
   * Choose the destination folder
   * Click "Move"


**Try it now:**
1. Open the **Workspace** browser in the sidebar
2. Navigate to your Home folder
3. Create a folder called "Practice Queries"
4. Save or move a query into that folder

---

**Tip:** You can organize queries alongside notebooks and other objects in the same folder structure which is great for keeping entire projects together
