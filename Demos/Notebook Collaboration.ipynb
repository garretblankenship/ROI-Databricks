````python
# Data Processing Function
# Try highlighting different parts of this code and adding comments!

import pandas as pd
from datetime import datetime, timedelta

def process_sales_data(df, days=30):
    """
    Process sales data for the last N days.
    
    Args:
        df: DataFrame with sales data
        days: Number of days to include (default: 30)
    
    Returns:
        Processed DataFrame with aggregated metrics
    """
    # Calculate date threshold
    cutoff_date = datetime.now() - timedelta(days=days)
    
    # Filter recent data
    recent_sales = df[df['date'] >= cutoff_date]
    
    # Aggregate by product
    summary = recent_sales.groupby('product').agg({
        'quantity': 'sum',
        'revenue': 'sum',
        'customer_id': 'nunique'
    }).reset_index()
    
    # Rename columns
    summary.columns = ['product', 'total_quantity', 'total_revenue', 'unique_customers']
    
    return summary

# Example: Highlight the date filtering logic and add a comment asking
# "Should we make the date column name configurable?"
````

## 4. Notebook Permissions and Sharing

Control who can access, view, run, and edit your notebooks with granular permission levels.

### ðŸ” **Five Permission Levels**

| Permission | View | Comment | Run via %run | Attach & Run | Edit | Manage Permissions |
|------------|------|---------|--------------|--------------|------|--------------------|
| **NO PERMISSIONS** | âŒ | âŒ | âŒ | âŒ | âŒ | âŒ |
| **CAN READ** | âœ… | âœ… | âœ… | âŒ | âŒ | âŒ |
| **CAN RUN** | âœ… | âœ… | âœ… | âœ… | âŒ | âŒ |
| **CAN EDIT** | âœ… | âœ… | âœ… | âœ… | âœ… | âŒ |
| **CAN MANAGE** | âœ… | âœ… | âœ… | âœ… | âœ… | âœ… |

### ðŸ“ **Permission Details**

**CAN READ**
* View all cells and their content
* Add comments to code
* Use notebook in `%run` commands or workflows
* Cannot execute cells or make changes

**CAN RUN**
* Everything in CAN READ, plus:
* Attach notebook to compute
* Run cells and see results
* Cannot modify code

**CAN EDIT**
* Everything in CAN RUN, plus:
* Modify cells and code
* Add/delete cells
* Save changes
* Cannot change permissions

**CAN MANAGE**
* Everything in CAN EDIT, plus:
* Modify permissions for other users
* Delete the notebook
* Full administrative control

### ðŸ‘¥ **Sharing a Notebook**

**Step-by-step:**

1. Click the **Share** button (top right of notebook)
2. Enter user email or group name
3. Select permission level from dropdown
4. Click **Add**
5. Click **Done**

**Tips:**
* Share with groups for easier management
* Use CAN READ for stakeholders who just need visibility
* Use CAN RUN for analysts who need to execute but not modify
* Use CAN EDIT for active collaborators
* Keep CAN MANAGE limited to notebook owners

### ðŸ›¡ï¸ **Security Best Practices**

âœ… **Do:**
* Review permissions regularly
* Use groups instead of individual users when possible
* Grant minimum necessary permissions
* Document who has access and why
* Remove access when team members change roles

âŒ **Don't:**
* Give CAN MANAGE to everyone
* Share sensitive notebooks publicly
* Run code in notebooks you don't trust
* Forget to audit permissions periodically

### âš ï¸ **Important Security Note**

**Only run code in notebooks you trust!**

When you run someone else's notebook:
* Code executes with YOUR credentials
* Can access YOUR data and resources
* Could potentially be malicious

Always review code before running, especially from external sources.

### ðŸ›ï¸ **Default Permissions**

**Workspace Admins:**
* Automatically have CAN MANAGE on all notebooks
* Can access any notebook in the workspace
* Can modify any permissions

**Notebook Creators:**
* Automatically have CAN MANAGE on notebooks they create
* Can share with others
* Can transfer ownership if needed

**Other Users:**
* Have NO PERMISSIONS by default
* Must be explicitly granted access
* Permissions can be inherited from folder permissions

### ðŸ“ **Folder-Level Permissions**

Permissions can be set at the folder level:
* All notebooks in folder inherit permissions
* Easier to manage team access
* Can override at individual notebook level
* Useful for organizing team workspaces
