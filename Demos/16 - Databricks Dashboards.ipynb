## What are Databricks Dashboards?

**Databricks Dashboards** (Lakeview) are interactive, visual reports built on your lakehouse data.

**Key features:**
* **Visual analytics** - Charts, graphs, counters
* **AI-powered** - Generate visualizations with natural language
* **Interactive filters** - Cross-filtering, global filters
* **Auto-refresh** - Scheduled updates
* **Multiple data sources** - Combine different tables
* **Custom calculations** - Create calculated fields
* **AI forecasting** - Predict future trends

### What is actually happening under the hood?

When you click **"Clone with AI Forecast"**, the dashboard creates a **new dataset** that wraps your original query with Databricks' built-in `ai_forecast()` SQL table-valued function. This is a real SQL function you can also use directly.

#### `ai_forecast()` Syntax

```sql
ai_forecast(
  observed       TABLE,                          -- Your historical data
  horizon        DATE | TIMESTAMP | STRING,      -- How far to predict
  time_col       STRING,                          -- Name of the time column
  value_col      STRING | ARRAY<STRING>,          -- Column(s) to forecast
  group_col      STRING | ARRAY<STRING> | NULL DEFAULT NULL,
  prediction_interval_width DOUBLE DEFAULT 0.95,  -- Confidence band (0 to 1)
  frequency      STRING DEFAULT 'auto',           -- 'day', 'week', 'month', etc.
  seed           INTEGER | NULL DEFAULT NULL,      -- For reproducibility
  parameters     STRING DEFAULT '{}'              -- JSON model config
)
```

#### What the dashboard generates

Behind the scenes, the cloned dataset query looks roughly like this:

```sql
WITH historical AS (
  SELECT
    DATE_TRUNC('week', created_at) AS booking_week,
    SUM(total_amount) AS weekly_revenue
  FROM samples.wanderbricks.bookings
  GROUP BY 1
)
SELECT * FROM ai_forecast(
  TABLE(historical),
  horizon  => '2025-06-01',        -- Forecast end date
  time_col => 'booking_week',
  value_col => 'weekly_revenue',
  prediction_interval_width => 0.95
)
```

#### Output columns

The function returns **only the future rows** (not the historical data) with these columns:

| Column | Description |
| --- | --- |
| `booking_week` | The future time period |
| `weekly_revenue_forecast` | The predicted value |
| `weekly_revenue_upper` | Upper bound of the confidence interval |
| `weekly_revenue_lower` | Lower bound of the confidence interval |

The dashboard then overlays these forecast rows on top of the historical line chart.

---

**Forecast works best with:**
* At least 2-3 months of historical data
* Regular patterns or trends
* Daily or weekly aggregations
* Numeric values (revenue, counts)

**Checkpoint:** Line chart with AI forecast showing future predictions

---

**Tip:** You can UNION the historical data with the forecast output to display both in a single chart:
```sql
SELECT booking_week, weekly_revenue AS value, 'actual' AS series FROM historical
UNION ALL
SELECT booking_week, weekly_revenue_forecast AS value, 'forecast' AS series
FROM ai_forecast(TABLE(historical), ...)
```

## Part 4: Interactive Filtering

Filters make dashboards interactive by letting users focus on specific data.

**Filter types:**
* **Cross-filtering** - Click a chart to filter others
* **Global filters** - Apply to entire dashboard
* **Page filters** - Apply to current page only
* **Widget filters** - Apply to specific visualization

---

**Tip:** Widget filters are "hidden" meaning viewers don't see them but data is filtered

## Part 5: Parameterized Datasets

Parameters in datasets make dashboards dynamic and interactive.

**What are dataset parameters?**
* Variables in dataset SQL queries
* Users can change values via widgets
* Filter data at the query level
* More efficient than post-query filtering

---

### Step 13: Create a Dataset with Parameters

**Follow these steps:**

1. Go to the **"Data"** tab
2. Click **"+ Add"** to create a new dataset
3. **Name:** "bookings_by_date_range"

4. **Enter this SQL query:**
```sql
SELECT 
  b.booking_id,
  b.total_amount,
  b.guests_count,
  b.created_at,
  DATE(b.created_at) AS booking_date,
  p.property_type,
  p.title AS property_name,
  CONCAT(u.name, ' (', u.country, ')') AS customer_info
FROM samples.wanderbricks.bookings b
JOIN samples.wanderbricks.properties p ON b.property_id = p.property_id
JOIN samples.wanderbricks.users u ON b.user_id = u.user_id
WHERE DATE(b.created_at) BETWEEN :date_range.start AND :date_range.end
  AND b.status = 'confirmed'
```

5. Notice the **`:date_range.start`** and **`:date_range.end`** syntax
   * This creates a date range parameter
   * `:parameter_name` is the parameter syntax
   * `.start` and `.end` for date ranges

6. Click **"Save"**

7. A **parameter widget** is automatically created!

**Parameter syntax:**
* `:param_name` - Single value parameter
* `:date_range.start` and `:date_range.end` - Date range
* `:param_name.min` and `:param_name.max` - Numeric range

✅ **Checkpoint:** Dataset created with date range parameter

---

**💡 Tip:** Parameters filter data at the source - more efficient than filtering after query

### ⚠️ Step 15: Understanding Parameters vs Filters

**Important difference:**

**Cross-filtering limitation with parameters:**

1. **Create two visualizations:**
   * Viz 1: From "bookings" dataset (no parameters)
   * Viz 2: From "bookings_by_date_range" dataset (with parameters)

2. **Try cross-filtering:**
   * Click on Viz 1
   * Viz 2 does NOT filter
   * **Why?** Different datasets = no cross-filtering

3. **But global/page filters still work:**
   * Add a global filter for property_type
   * Both visualizations filter correctly
   * Global filters work across datasets

**Key differences:**

| Feature | Parameters | Filters |
|---------|------------|----------|
| **Location** | In SQL query | After query runs |
| **Efficiency** | More efficient | Less efficient |
| **Cross-filtering** | Breaks it (different datasets) | Works |
| **Global filters** | Still work | Works |
| **Use case** | Date ranges, thresholds | Interactive exploration |

**When to use each:**
* **Parameters:** Date ranges, large data filtering, performance
* **Filters:** Interactive exploration, cross-filtering, drill-down

**Best practice:**
* Use parameters for date ranges (efficient)
* Use same non-parameterized dataset for cross-filtering
* Use global filters for common dimensions

✅ **Checkpoint:** Understand parameter limitations with cross-filtering

---

**💡 Tip:** For cross-filtering, use the same dataset; for efficiency, use parameters

3. **Calculation 3:** `lead_time_category`
   ```sql
   CASE
     WHEN DATEDIFF(check_in, created_at) <= 7 THEN 'Last Minute'
     WHEN DATEDIFF(check_in, created_at) <= 30 THEN 'Short Notice'
     WHEN DATEDIFF(check_in, created_at) <= 90 THEN 'Advance'
     ELSE 'Early Bird'
   END
   ```

**Part B: Create visualizations**

1. **Visualization 1:** Bar chart
   * X-axis: lead_time_category
   * Y-axis: COUNT of bookings
   * Title: "Bookings by Lead Time"

2. **Visualization 2:** Scatter plot
   * X-axis: booking_lead_time
   * Y-axis: total_amount
   * Title: "Booking Value vs Lead Time"

**Expected result:**
* Three new calculated fields
* Two visualizations using them
* Insights into booking patterns

## 🔧 Troubleshooting Common Issues

**Visualization not showing data:**
* ✅ Check dataset query returns data
* ✅ Verify column mappings are correct
* ✅ Check filters aren't excluding all data
* ✅ Refresh the dataset
* ✅ Check SQL warehouse is running

**Cross-filtering not working:**
* ✅ Verify both visualizations use **same dataset**
* ✅ Check cross-filtering is enabled
* ✅ Ensure common column exists
* ✅ Try clicking different parts of the chart

**Parameters not updating:**
* ✅ Check parameter syntax: `:param_name`
* ✅ Verify parameter widget is visible
* ✅ Check dataset query is correct
* ✅ Try refreshing the dashboard

**Calculated field errors:**
* ✅ Check formula syntax
* ✅ Verify column names are correct
* ✅ Test formula in SQL query first
* ✅ Check for NULL handling

**Schedule not running:**
* ✅ Verify schedule is enabled
* ✅ Check SQL warehouse is available
* ✅ Review refresh history for errors
* ✅ Verify permissions on tables

**Dashboard slow to load:**
* ✅ Reduce number of visualizations
* ✅ Add LIMIT to dataset queries
* ✅ Use aggregations in queries
* ✅ Increase SQL warehouse size
* ✅ Check for expensive calculations

**AI features not available:**
* ✅ Check workspace has AI features enabled
* ✅ Verify you have appropriate permissions
* ✅ Ensure sufficient data for forecasting
