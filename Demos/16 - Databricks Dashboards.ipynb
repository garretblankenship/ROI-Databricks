## What are Databricks Dashboards?

**Databricks Dashboards** (Lakeview) are interactive, visual reports built on your lakehouse data.

**Key features:**
* **Visual analytics** - Charts, graphs, counters
* **AI-powered** - Generate visualizations with natural language
* **Interactive filters** - Cross-filtering, global filters
* **Auto-refresh** - Scheduled updates
* **Multiple data sources** - Combine different tables
* **Custom calculations** - Create calculated fields
* **AI forecasting** - Predict future trends

### What is actually happening under the hood?

When you click **"Clone with AI Forecast"**, the dashboard creates a **new dataset** that wraps your original query with Databricks' built-in `ai_forecast()` SQL table-valued function. This is a real SQL function you can also use directly.

#### `ai_forecast()` Syntax

```sql
ai_forecast(
  observed       TABLE,                          -- Your historical data
  horizon        DATE | TIMESTAMP | STRING,      -- How far to predict
  time_col       STRING,                          -- Name of the time column
  value_col      STRING | ARRAY<STRING>,          -- Column(s) to forecast
  group_col      STRING | ARRAY<STRING> | NULL DEFAULT NULL,
  prediction_interval_width DOUBLE DEFAULT 0.95,  -- Confidence band (0 to 1)
  frequency      STRING DEFAULT 'auto',           -- 'day', 'week', 'month', etc.
  seed           INTEGER | NULL DEFAULT NULL,      -- For reproducibility
  parameters     STRING DEFAULT '{}'              -- JSON model config
)
```

#### What the dashboard generates

Behind the scenes, the cloned dataset query looks roughly like this:

```sql
WITH historical AS (
  SELECT
    DATE_TRUNC('week', created_at) AS booking_week,
    SUM(total_amount) AS weekly_revenue
  FROM samples.wanderbricks.bookings
  GROUP BY 1
)
SELECT * FROM ai_forecast(
  TABLE(historical),
  horizon  => '2025-06-01',        -- Forecast end date
  time_col => 'booking_week',
  value_col => 'weekly_revenue',
  prediction_interval_width => 0.95
)
```

#### Output columns

The function returns **only the future rows** (not the historical data) with these columns:

| Column | Description |
| --- | --- |
| `booking_week` | The future time period |
| `weekly_revenue_forecast` | The predicted value |
| `weekly_revenue_upper` | Upper bound of the confidence interval |
| `weekly_revenue_lower` | Lower bound of the confidence interval |

The dashboard then overlays these forecast rows on top of the historical line chart.

---

**Forecast works best with:**
* At least 2-3 months of historical data
* Regular patterns or trends
* Daily or weekly aggregations
* Numeric values (revenue, counts)

**Checkpoint:** Line chart with AI forecast showing future predictions

---

**Tip:** You can UNION the historical data with the forecast output to display both in a single chart:
```sql
SELECT booking_week, weekly_revenue AS value, 'actual' AS series FROM historical
UNION ALL
SELECT booking_week, weekly_revenue_forecast AS value, 'forecast' AS series
FROM ai_forecast(TABLE(historical), ...)
```

## Part 4: Interactive Filtering

Filters make dashboards interactive by letting users focus on specific data.

**Filter types:**
* **Cross-filtering** - Click a chart to filter others
* **Global filters** - Apply to entire dashboard
* **Page filters** - Apply to current page only
* **Widget filters** - Apply to specific visualization

---

| Example | Date range, region | Specific metric filter |

```
Widget Filter (most specific)
    ↓
Page Filter
    ↓
Global Filter (least specific)
```

All filters combine (AND logic).

**Tip:** Widget filters are "hidden" meaning viewers don't see them but data is filtered

## Part 5: Parameterized Datasets

Parameters in datasets make dashboards dynamic and interactive.

**What are dataset parameters?**
* Variables in dataset SQL queries
* Users can change values via widgets
* Filter data at the query level
* More efficient than post-query filtering

---

### Step 13: Create a Dataset with Parameters

**Follow these steps:**

1. Go to the **"Data"** tab
2. Click **"Create from SQL** to create a new dataset
3. **Name:** "bookings_by_date_range"

4. **Enter this SQL query:**
```sql
SELECT 
  b.booking_id,
  b.total_amount,
  b.guests_count,
  b.created_at,
  DATE(b.created_at) AS booking_date,
  p.property_type,
  p.title AS property_name,
  CONCAT(u.name, ' (', u.country, ')') AS customer_info
FROM samples.wanderbricks.bookings b
JOIN samples.wanderbricks.properties p ON b.property_id = p.property_id
JOIN samples.wanderbricks.users u ON b.user_id = u.user_id
WHERE DATE(b.created_at) BETWEEN :date_range.min AND :date_range.max
  AND b.status = 'confirmed'
```

5. Notice the **`:date_range.min`** and **`:date_range.max`** syntax
   * This creates a date range parameter
   * `:parameter_name` is the parameter syntax
   * `.min` and `.max` for date ranges

6. A **parameter widget** is automatically created!

7. Run the query so the Dashboard can understand the datasource schema

**Checkpoint:** Dataset created with date range parameter

---

**Tip:** Parameters filter data at the source more efficient than filtering after query

## Part 6: Custom Calculations

**Custom calculations** let you create new fields using formulas without modifying the dataset query. Custom calculations are computed at the widget/visualization level, not as part of the SQL query itself. 

**What are custom calculations?**
* Calculated fields created in the dashboard
* Use existing columns in formulas
* Support SQL expressions
* Reusable across visualizations

**Use cases:**
* Mathematical calculations
* Conditional logic (CASE statements)
* String concatenation
* Date calculations
* Business metrics

---

### Step 16: Create a Simple Calculated Field

**Follow these steps:**

1. Go to the **"Data"** tab
2. Select a dataset (e.g., "bookings")
3. Look for and click the **"Add custom calculation"** button

**Create first calculation:**

4. **Calculation settings:**
   * **Name:** `revenue_per_night`
   * **Formula:**
   ```sql
   total_amount / nights
   ```
   * **Comment:** "Average revenue per night"

5. Click **"Create"**

6. The new field appears in the datasets schema but not in the query results!

If you want to see revenue_per_night in the query results table directly, you would need to add total_amount / DATEDIFF(check_out, check_in) AS revenue_per_night to the SQL SELECT. However, keeping it as a custom calculation is the recommended approach since it's easier to maintain and keeps your SQL simple.

**Using the calculated field:**

7. Create a new visualization:
   * Dataset: "bookings"
   * Type: Bar chart
   * X-axis: booking_month
   * Y-axis: AVG of `revenue_per_night` (your calculated field)
   * Title: "Average Revenue per Night by Month"

8. The calculated field works like any other column!

**Checkpoint:** Calculated field created and used in visualization

---

**Tip:** Calculated fields are computed on-the-fly from the dataset

### Step 17: Create a CASE Statement Calculation

3. **Calculation settings:**

### Step 19: Publish and Share Your Dashboard

**What is publishing?**
* Makes your dashboard available to viewers
* Switches from draft (edit) mode to a published version
* Requires choosing a **data permission model** that controls how viewers access the underlying data

**Data permission options when publishing:**

| Option | Who runs queries | Data access determined by | Best for |
|--------|-----------------|--------------------------|----------|
| **Share data permissions** (default) | Publisher | Publisher's permissions | Consistent experience for users without direct data access |
| **Individual data permissions** | Each viewer | Viewer's own permissions | Internal users with workspace accounts and data permissions |

* **Share data permissions** lets viewers see the dashboard even without direct access to the underlying data or compute. This is the default.
* **Individual data permissions** requires each viewer to have their own access to the data. Compute access is still granted by the publisher's credentials.
* For **draft** dashboards, the viewer's own data permissions are always applied, regardless of the publishing setting.

**To publish:**

1. Click the **"Publish"** button (top-right)
2. Choose your data permission setting
3. Click **"Publish"** to confirm

**To share a published dashboard:**

1. Click the **"Share"** button
2. Use the **Sharing dialog** to grant access:
   * Enter users or groups and assign permission levels (**CAN EDIT**, **CAN MANAGE**), then click **Add**
   * Use **Sharing settings** at the bottom to quickly grant view access to anyone in your Databricks account
3. Click **Copy link** to get a shareable URL for the published dashboard
4. Optionally, **embed** the dashboard in an external website or application

**Dashboard permission levels (ACLs):**

| Permission | View & interact | Refresh | Edit | Publish | Clone | Modify permissions | Delete |
|------------|:-:|:-:|:-:|:-:|:-:|:-:|:-:|
| **CAN VIEW / CAN RUN** | ✓ | ✓ | | | ✓ | | |
| **CAN EDIT** | ✓ | ✓ | ✓ | ✓ | ✓ | | |
| **CAN MANAGE** | ✓ | ✓ | ✓ | ✓ | ✓ | ✓ | ✓ |

> **Note:** Users who do not have access to the workspace are limited to **CAN RUN** permissions, even if granted a higher level. Elevated permissions take effect only after the user is added to the workspace.

**Published vs Draft:**

| Mode | Purpose | Who sees it | Can edit |
|------|---------|-------------|----------|
| **Draft** | Building & editing | Only users with CAN EDIT or higher | Yes |
| **Published** | Viewing | Everyone with granted access | No (must enter edit mode) |

**After publishing:**
* Click **"View Published"** to see how everyone will view your dasbhoard
* Click **"Edit draft"** to go back to edit mode to make changes (saved as a new draft)
* Click **"Publish"** again to push draft changes to the published version
* You can also publish dashboards via the **REST API**, including with **service principal credentials** for automated workflows

**Checkpoint:** Dashboard is published with appropriate data permissions and shared with viewers

---

**Tip:** Use "Share data permissions" when sharing with users who don't have direct data access. Use "Individual data permissions" for internal teams who should only see data they're authorized for.

   
### Step 20: Schedule Automatic Refresh & Manage Subscriptions

Scheduling dashboard updates ensures dashboards display up-to-date information and improves performance for end users. **Subscriptions** keep stakeholders informed by sending dashboard snapshots to **email**, **Slack**, or **Microsoft Teams**.

---

**To create a schedule:**

1. Make sure you are viewing the published Dashboard and in the upper-right corner of the dashboard, click **"Schedule"**
2. Enter a **name** identifier for your new schedule
3. Use the **drop-down selectors** to specify the frequency and start time
   * Optionally, select **"Show cron syntax"** to edit the schedule using Quartz Cron Syntax
4. Use the **Timezone** drop-down to select the time zone
5. Choose how filter values are applied during scheduled runs:
   * **Use default values** — applies the default values from the dashboard's filter widgets
   * **Use filter selections** — applies the filter values that were set when the schedule was created
6. Use the **SQL warehouse** drop-down to choose a warehouse for scheduled runs (defaults to the warehouse used to create the dashboard)
7. *(Optional)* Click **"Subscribers"** to add subscribers (see below)
8. Click **Create**

---

**Subscriptions:**

Subscribers receive a **snapshot** of the dashboard each time the schedule runs. For dashboards with multiple pages, the snapshot shows the **first page**.

**Supported subscription destinations:**

| Destination | Format delivered | Requirements |
|-------------|-----------------|---|
| **Email** | PDF snapshot via email | Workspace users or configured email notification destinations (distribution lists, external users) |
| **Slack** | PNG image in channel + direct link to dashboard + PDF in message thread | Workspace admin must configure Slack notification destinations |
| **Microsoft Teams** | PNG image in channel + direct link to dashboard + PDF in message thread | Workspace admin must configure Teams notification destinations |

**Who can manage subscribers?**
* Users with **CAN EDIT** or higher can add and remove subscribers
* Users with **CAN VIEW** can add and remove **themselves** as a subscriber to an existing schedule

> **Note:** If you cannot add email subscribers, your workspace admin may have turned off email notifications.

---

**Cache benefits of scheduled updates:**

* All scheduled updates trigger **cache refreshes** that improve dashboard load times for all viewers
* Dashboards published with **shared data permissions** populate a shared cache accessible by all viewers
* With **individual data permissions**, users can opt for **refresh-only schedules** (faster cache without notifications) or subscribe to also receive snapshots
* Cached results mean queries run less frequently, **reducing compute costs**

---

**Best practices:**
* Match refresh frequency to your data update cadence
* Schedule during off-peak hours to reduce warehouse load
* Use refresh-only schedules for cost-conscious teams that still want fast load times
* Monitor refresh history to catch failures early

**Checkpoint:** Dashboard schedule created with appropriate subscribers

---

**Tip:** Subscriptions support email, Slack, and Teams — choose the channel where your stakeholders already work

   
### Step 21: Organize Your Dashboards

Dashboards are managed in two places: the **Workspace browser** for organizing into folders, and the **Dashboards listing page** for searching and filtering.

---

**Using folders (Workspace browser):**

1. Click **Workspace** in the sidebar to open the workspace browser
2. Dashboards are stored in `/Workspace/Users/<your-username>` by default
3. Click **Create** > **Folder** to create a new folder (e.g., "Booking Analytics", "Executive Reports")
4. **Move a dashboard to a folder:**
   * Right-click the dashboard (or click the kebab menu)
   * Select **"Move"**
   * Choose the destination folder
   * Click **"Move"**
5. **Set permissions at the folder level:**
   * Right-click the folder and select **"Share"**
   * Enter users, groups, or service principals
   * Select the permission level and click **Add**

---

**Using the Dashboards listing page:**

1. Click **Dashboards** in the sidebar to open the listing page
2. The listing page shows dashboards you have access to, sorted in reverse-chronological order
3. **Search:** Type in the search bar to find dashboards by name, page names, widget titles, descriptions, or dataset names
4. **Filter:** Use the filter options to narrow results by:
   * **Owner** — search by the asset owner
   * **Status** — show only **Certified** assets or your **Favorites**
   * **Last modified** — filter by time period

> **Note:** The Dashboards listing page is for **viewing and finding** dashboards. To organize dashboards into folders, use the **Workspace browser**.

---

**Using favorites:**

1. Click the **star icon** next to the dashboard name (available in both the Workspace browser and the Dashboards listing page)
2. The dashboard is marked as a favorite
3. Filter by **Favorites** on the listing page for quick access

---

**Organization best practices:**
* Use workspace folders to group dashboards by department or project
* Set folder-level permissions to share all dashboards within a folder at once
* Star frequently used dashboards for quick access
* Use clear, descriptive dashboard names (searchable from the listing page)
* Request **Certified** status for production-ready dashboards

**Checkpoint:** Dashboard is organized in a workspace folder and easy to find

## Troubleshooting Common Issues

**Visualization not showing data:**
* Check dataset query returns data
* Verify column mappings are correct
* Check filters aren't excluding all data
* Refresh the dataset
* Check SQL warehouse is running

**Cross-filtering not working:**
* Verify both visualizations use **same dataset**
* Check cross-filtering is enabled
* Ensure common column exists
* Try clicking different parts of the chart

**Parameters not updating:**
* Check parameter syntax: `:param_name`
* Verify parameter widget is visible
* Check dataset query is correct
* Try refreshing the dashboard

**Calculated field errors:**
* Check formula syntax
* Verify column names are correct
* Test formula in SQL query first
* Check for NULL handling

**Schedule not running:**
* Verify schedule is enabled
* Check SQL warehouse is available
* Review refresh history for errors
* Verify permissions on tables

**Dashboard slow to load:**
* Reduce number of visualizations
* Add LIMIT to dataset queries
* Use aggregations in queries
* Increase SQL warehouse size
* Check for expensive calculations

**AI features not available:**
* Check workspace has AI features enabled
* Verify you have appropriate permissions
* Ensure sufficient data for forecasting
