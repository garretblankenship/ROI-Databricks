# Navigating Unity Catalog Via The UI

## Instructions


## Overview

This hands-on demo guides you through exploring Unity Catalog's three-level namespace (Catalog → Schema → Table) using the Databricks UI. You'll learn how to navigate the Catalog explorer, inspect metadata at each level, and preview sample data—essential skills for understanding your data organization and governance structure.

**What you'll explore:**
* Catalog browser navigation and hierarchy
* Metadata inspection (Overview and Details tabs)
* Sample data preview for tables
* The `samples.nyctaxi.trips` dataset as a practical example

1. In the Left Sidbar click on `Catalog`
![alt text](./Assets/11/a.png)

2. Review all of the Catalogs, Schemas and Data Objects Available.
3. Click on the `samples` Catalog
![alt text](./Assets/11/b.png)

4. Review the `Overview` and `Details` tab for this Catalog.
5. Click on the `nyctaxi` Schema.
![alt text](./Assets/11/c.png)

6. Review the `Overview` and `Details` tab for this Schema.
7. Click the `trips` Table.
![alt text](./Assets/11/d.png)


8. Review the `Overview`, `Sample Data` and `Details` tab for this Table.
