## üßû What is Genie?

**Genie** is Databricks' AI-powered analytics assistant that understands natural language questions.

**Key features:**
* üí¨ **Natural language** - Ask questions in plain English
* ü§ñ **AI-generated SQL** - Automatically writes queries
* üìä **Auto-visualization** - Creates appropriate charts
* üß† **Learns from feedback** - Improves with benchmarks
* üîó **Semantic understanding** - Knows your data context
* üìà **Conversational** - Follow-up questions

**Genie vs Traditional Dashboards:**

| Feature | Dashboards | Genie Spaces |
|---------|------------|---------------|
| **Query method** | Write SQL | Ask questions |
| **Visualization** | Manual config | AI-generated |
| **Best for** | Structured reports | Ad-hoc exploration |
| **Users** | SQL knowledge needed | No SQL needed |
| **Flexibility** | Fixed layout | Conversational |

### ‚ûï Step 15: Add a Benchmark Manually

```sql
SELECT 
  YEAR(r.created_at) AS year,
  MONTH(r.created_at) AS month,
  p.property_type,
  AVG(r.rating) AS avg_rating
FROM samples.wanderbricks.reviews r
JOIN samples.wanderbricks.properties p 
  ON r.property_id = p.property_id
WHERE r.is_deleted = false
GROUP BY YEAR(r.created_at), MONTH(r.created_at), p.property_type
ORDER BY year, month, property_type
```

### üéØ **Challenge 2: Create Multiple Benchmarks**

**Your task:**

Create 3 more benchmarks covering different question types.

**Benchmark 1: Simple aggregation**
* **Question:** "How many confirmed bookings do we have?"
* **SQL:**
```sql
SELECT COUNT(*) AS booking_count
FROM samples.wanderbricks.bookings
WHERE status = 'confirmed'
```

**Benchmark 2: Time-series analysis**
* **Question:** "Show me daily booking revenue for the last 30 days"
* **SQL:**
```sql
SELECT 
  DATE(created_at) AS booking_date,
  SUM(total_amount) AS daily_revenue
FROM samples.wanderbricks.bookings
WHERE status = 'confirmed'
  AND created_at >= CURRENT_DATE() - INTERVAL 30 DAYS
GROUP BY DATE(created_at)
ORDER BY booking_date
```

**Benchmark 3: Multi-table join**
* **Question:** "Which properties have the highest average review ratings?"
* **SQL:**
```sql
SELECT 
  p.title AS property_name,
  p.property_type,
  AVG(r.rating) AS avg_rating,
  COUNT(r.review_id) AS review_count
FROM samples.wanderbricks.properties p
JOIN samples.wanderbricks.reviews r 
  ON p.property_id = r.property_id
WHERE r.is_deleted = false
GROUP BY p.title, p.property_type
HAVING COUNT(r.review_id) >= 5
ORDER BY avg_rating DESC
LIMIT 10
```
