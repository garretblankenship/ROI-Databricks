## 1. Query Parameters (Widgets)

**Query parameters** make your queries dynamic and interactive by allowing users to provide input values.

**What are parameters?**
* Input fields that appear above your query
* Users can change values without editing SQL
* Perfect for filters, date ranges, thresholds
* Also called "widgets" or "filters"

**Use cases:**
* Date range selection
* Location filters
*Threshold values
* Category selection
* Dashboard interactivity

---

   
### Step 1: Create a Text Parameter

**Try this:**

1. In SQL Editor, write this query:
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE pickup_zip = :zip_code
GROUP BY pickup_zip
```

2. Notice the `:zip_code` syntax - this creates a parameter!

3. When you type `:` followed by a name, a **parameter widget** appears belows the query

4. The widget shows:
   * Parameter name: "zip_code"
   * Input field (empty by default)
   * Gear icon for settings

**Try it now:**
1. Type the query above
2. See the parameter widget appear
3. Enter a zip code (e.g., "10103")
4. Run the query

---

**Tip:** Use `:parameter_name` syntax anywhere in your SQL

### Step 2: Configure Parameter Settings

**Try this:**

1. Click the **gear icon** on the parameter widget
2. A settings dialog opens with options
3. Configure your parameter:
   * Parameter Name: "zip_code"
   * Widget Label: "Zip Code"
   * Paramter Type: "Integer"
   * Widget Type: Text
   * Default value: "10103"

4. Click **"Save"** (over in the top right of the editor)

5. The widget now shows "Zip Code" as the label

---

**Tip:** Good titles make parameters user-friendly for dashboard viewers

   
   
### Step 3: Different Parameter Types

**Databricks Queries supports multiple parameter types:**

**1. Text**
* Free-form text input
* Use for: IDs, names, search terms
* Example: `:customer_name`

**2. Dropdown**
* Select from predefined options
* Use for: Categories, statuses, regions
* Example: `:status`

**3. Combobox**
* Select from predefined options or type a custom value
* Use for: Flexible lists where users may need values beyond the preset choices
* Example: `:city`

**4. Multiselect**
* Select multiple values from predefined options
* Use for: Filtering by several categories at once
* Example: `:selected_zips`

**5. Date Range**
* Start and end date pickers
* Use for: Time period analysis
* Example: `:date_range.min` and `:date_range.max`

**6. Dynamic Dropdown**
* Options populated from another query
* Use for: Dynamic lists
* Example: List of active customers

---

   
### Step 4: Create a Dropdown Parameter

**Try this:**

Create a query with a dropdown for trip status:

1. Write this query:
```sql
SELECT 
  DATE(tpep_pickup_datetime) AS trip_date,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE fare_amount > :min_fare
GROUP BY DATE(tpep_pickup_datetime)
ORDER BY trip_date DESC
LIMIT 30
```

2. A parameter widget appears for `min_fare`

3. Click the **gear icon**

4. Configure:
   * Widget Label: "Minimum Fare"
   * Widget Type: **Dropdown**
   * Parameter Type: **Integer**
   * Values: Add options:
     - `10` 
     - `20`
     - `50` 
     - `100`
   * Default: `10`

5. Click **"Save"**

6. The widget now shows a dropdown menu

---

**Tip:** Dropdowns prevent invalid input and guide users to valid options

   
### Step 5: Create a Date Parameter

**Try this:**

Create a query with a date filter:

1. Write this query:
```sql
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count,
  SUM(fare_amount) AS total_revenue
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) = :trip_date
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10
```

2. Click the gear icon on the `trip_date` parameter

3. Configure:
   * Widget Label: "Trip Date"
   * Parameter Type: **Date**
   * Widget Type: **Text**
   * Default value: Select a date (e.g., 2016-02-17)

4. Click **"Save"**

5. A **date picker** widget appears

---

**Tip:** Date parameters are essential for time-series analysis and reports

   
### Step 6: Use Multiple Parameters Together

**Try this:**

Create a query with multiple parameters:

1. Write this query:
```sql
SELECT 
  pickup_zip,
  dropoff_zip,
  COUNT(*) AS trip_count,
  AVG(fare_amount) AS avg_fare,
  AVG(trip_distance) AS avg_distance
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= :start_date
  AND DATE(tpep_pickup_datetime) <= :end_date
  AND fare_amount >= :min_fare
  AND trip_distance >= :min_distance
GROUP BY pickup_zip, dropoff_zip
ORDER BY trip_count DESC
LIMIT :result_limit
```

2. **Five parameters** appear:
   * `start_date` - Date
   * `end_date` - Date
   * `min_fare` - Number
   * `min_distance` - Number
   * `result_limit` - Number

3. Configure each parameter with appropriate types and defaults

4. All parameters appear in a row above the query

---

**Tip:** Multiple parameters create powerful, flexible queries for dashboards

## 2. Multiple Statement Results

**Multiple statement queries** let you run several SQL statements in one query and see all results.

**What is it?**
* Write multiple SQL statements separated by semicolons
* Run them all at once
* Each statement produces its own result set
* Results appear in separate tabs

**Use cases:**
* Data quality checks (multiple validation queries)
* Before/after comparisons
* Multi-step analysis
* Comprehensive reports

---

### Step 7: Write Multiple SQL Statements

**Try this:**

1. Write multiple statements separated by semicolons:

```sql
-- Statement 1: Overall statistics
SELECT 
  COUNT(*) AS total_trips,
  AVG(fare_amount) AS avg_fare,
  AVG(trip_distance) AS avg_distance
FROM samples.nyctaxi.trips
LIMIT 1;

-- Statement 2: Top pickup locations
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10;

-- Statement 3: Fare distribution
SELECT 
  CASE 
    WHEN fare_amount < 10 THEN 'Under $10'
    WHEN fare_amount < 20 THEN '$10-$20'
    WHEN fare_amount < 50 THEN '$20-$50'
    ELSE '$50+'
  END AS fare_range,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
GROUP BY fare_range
ORDER BY fare_range;
```

2. Notice the **semicolons** (`;`) separating statements

3. Each statement is independent

**Try it now:**
1. Copy the query above into SQL Editor
2. Notice you have 3 separate SELECT statements

---

**Tip:** Add comments to explain what each statement does

### Step 8: Run and View Multiple Results

**Try this:**

1. With your multi-statement query, click **"Run"** (or `Ctrl+Enter`)

2. All statements execute in sequence

3. The **Results panel** shows:
   * **Pages** for each statement (Result 1, Result 2, Result 3)
   * Click arrows to switch between results
   * Each page shows its own data table

4. Each result page displays:
   * Table with data
   * Row count
   * Execution time for that statement

---

**Tip:** Results are numbered in the order statements appear in your query

   
### Step 9: Combine Multiple Statements with Parameters

**Try this:**

Use parameters across multiple statements:

```sql
-- Statement 1: Summary for date range
SELECT 
  'Summary' AS report_section,
  COUNT(*) AS total_trips,
  SUM(fare_amount) AS total_revenue,
  AVG(fare_amount) AS avg_fare
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= :start_date
  AND DATE(tpep_pickup_datetime) <= :end_date;

-- Statement 2: Daily breakdown
SELECT 
  DATE(tpep_pickup_datetime) AS trip_date,
  COUNT(*) AS trip_count,
  SUM(fare_amount) AS daily_revenue
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= :start_date
  AND DATE(tpep_pickup_datetime) <= :end_date
GROUP BY DATE(tpep_pickup_datetime)
ORDER BY trip_date;

-- Statement 3: Top locations in period
SELECT 
  pickup_zip,
  COUNT(*) AS trip_count
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= :start_date
  AND DATE(tpep_pickup_datetime) <= :end_date
GROUP BY pickup_zip
ORDER BY trip_count DESC
LIMIT 10;
```

**What happens:**
* Same parameters (`:start_date`, `:end_date`) used in all statements
* Change parameters once, all statements update
* Creates a comprehensive report

## 4. Query Scheduling

**Query scheduling** lets you run queries automatically on a schedule.

**What is scheduling?**
* Queries run automatically at specified times
* No manual execution needed
* Results are cached and available
* Perfect for reports and dashboards

**Use cases:**
* Daily reports (run every morning)
* Hourly data refreshes
* Weekly summaries
* Real-time dashboard updates
* Automated data quality checks

---

**Available frequencies:**

**1. Never (Manual only)**
* Query only runs when you click Run
* No automatic execution
* Default for new queries

**2. Minutes**
* Every N minutes (e.g., every 5, 15, 30 minutes)
* Use for: Real-time dashboards, monitoring
* Example: Every 15 minutes

**3. Hourly**
* Every N hours (e.g., every 1, 2, 6 hours)
* Use for: Frequent updates, operational dashboards
* Example: Every hour at :00

**4. Daily**
* Every day at specific time
* Use for: Daily reports, morning summaries
* Example: Every day at 8:00 AM

**5. Weekly**
* Specific days of week at specific time
* Use for: Weekly reports, summaries
* Example: Every Monday at 9:00 AM

**6. Monthly**
* Specific day of month at specific time
* Use for: Monthly reports, end-of-month summaries
* Example: 1st of month at 8:00 AM

**7. Custom (Cron)**
* Advanced cron expression
* Use for: Complex schedules
* Example: `0 8 * * 1-5` (weekdays at 8 AM)

---

## 5. Alerts

**Alerts** notify you when query results meet specific conditions.

**What are alerts?**
* Self-contained monitors with their own query editor and schedule
* Automatically run a query on a schedule and evaluate a condition
* Send notifications when conditions are met
* Include an alert history to review past evaluations

**Use cases:**
* Data quality issues (null count > threshold)
* Business metrics (revenue drops below target)
* System monitoring (error count increases)
* Anomaly detection (unusual patterns)
* SLA monitoring (response time too high)

**Key components:**
1. **Query editor** - Write and test the alert query directly
2. **Schedule** - Set how often the alert runs
3. **Condition** - Define the threshold that triggers the alert
4. **Notifications** - Choose who gets notified

---

### Step 14: Create Your First Alert

**Try this:**

Create an alert for high-value taxi fare days:

1. Click **Alerts** in the sidebar and click **Create Alert**

2. In the **query editor**, paste the following query:
```sql
SELECT
  to_date(tpep_pickup_datetime) AS date,
  SUM(fare_amount) AS amount
FROM samples.nyctaxi.trips
GROUP BY ALL
ORDER BY 1 DESC
```

3. Click **Run all (1000)** to test the query and see results

4. In the **Condition** section, configure:
   * **Aggregation:** Select **Sum**
   * **Column:** Select **amount**
   * **Operator:** Select **>** (greater than)
   * **Threshold:** Select **Static value** and enter **4000**

5. Click **Test condition** to preview whether the alert would trigger with the current data

6. In the **Notifications** section:
   * Search for your user name
   * Click on it when it appears to add yourself as a subscriber

   **Notification behavior:**
      * Subscribers receive an email when the alert **triggers** (condition is met)
      * The notification includes:
      * Alert name
      * Threshold and actual value
      * Link to the alert
      * Timestamp

   **Notification destinations:**
   * You can also send alerts to configured **notification destinations** (e.g., Slack channels, PagerDuty, webhooks)
   * These are set up by your workspace admin under **Settings > Notification destinations**

7. Click the **Schedule Button** to open the **Edit schedule** dialog:
   * Set the alert to run **Every 5 minutes** from **0 minutes past the hour**
   * Optionally check **Show cron syntax** to edit in Quartz Cron format
   * Click **Save**

8. Click **View alert** to save and activate your alert

---

**Tip:** Alerts are self-contained, they include their own query, schedule, and conditions all in one place

> **Note:** During the Public Preview period, alerts do not support queries with parameters. Use [legacy alerts](https://docs.databricks.com/sql/user/alerts/index.html) if you need parameterized alert queries.

### Alert Condition Types

**How conditions work:**

Alert conditions evaluate a **data value** against a **threshold** using a **logical operator**.

**Data value options:**

You can set the condition on:
* **First value** of a column in the query result (single row)
* **An aggregation** across all rows of a single column:
  * **Sum** — Total of all values
  * **Average** — Mean of all values
  * **Count** — Number of rows
  * **Min** — Smallest value
  * **Max** — Largest value

**Logical operators:**

| Operator | Description | Example Use Case |
|----------|-------------|------------------|
| **>** (Greater than) | Triggers when value exceeds threshold | Error count > 100 |
| **>=** (Greater than or equal) | Triggers when value meets or exceeds | Revenue >= 10000 |
| **<** (Less than) | Triggers when value falls below threshold | Active users < 1000 |
| **<=** (Less than or equal) | Triggers when value meets or falls below | Inventory <= 10 |
| **=** (Equals) | Triggers when value exactly matches | Failed jobs = 0 |
| **!=** (Not equals) | Triggers when value doesn't match | Status != 0 |

**Threshold types:**
* **Static value** — A fixed number you enter (default)

**Example queries for alerts:**
```sql
-- Revenue monitoring
SELECT
  to_date(order_date) AS date,
  SUM(amount) AS daily_revenue
FROM sales.orders
GROUP BY ALL
ORDER BY 1 DESC

-- Error count monitoring
SELECT
  to_date(timestamp) AS date,
  COUNT(*) AS error_count
FROM system.errors
GROUP BY ALL
ORDER BY 1 DESC
```

> **Tip:** Use the **Test condition** button to preview the alert evaluation against the current data before saving.

## 6. Query Snippets

**Query snippets** are reusable segments of SQL code you can insert into any query.

**What are snippets?**
* Saved SQL code templates
* Can include insertion points with placeholder text for runtime values
* Shared across your workspace
* Work in the SQL editor, notebook SQL cells, SQL files, and AI/BI dashboard queries

**Use cases:**
* Frequent JOIN statements
* Complicated clauses like WITH or CASE
* Conditional formatting
* Common WHERE clauses
* Standard date filters

---

   
### Step 15: Create Your First Snippet

**Try this:**

Create a reusable date filter snippet:

1. Open the **Query snippets** panel:
   * In the **SQL editor**: click the **kebab menu** (⋮), then click **View**, then click **Query snippets**
   * In a **notebook** or **file editor**: click **View**, then click **Query snippets**

2. Click **Create query snippet**

3. A snippet editor opens:

4. Configure your snippet:
   * **Replace:** `datefilter` (the name you type to insert the snippet)
   * **Description:** "Standard date range WHERE clause" *(optional)*
   * **Snippet:**
   ```sql
   WHERE DATE(${1:date_column}) >= '${2:start_date}'
     AND DATE(${1:date_column}) <= '${3:end_date}'
   ```

5. Click **Create**

**Snippet syntax — insertion points:**
* `${1:placeholder}` — First insertion point with placeholder text
* `${2:placeholder}` — Second insertion point
* `${1:}` — Reference to first insertion point (reused)
* `${0:placeholder}` — Always the **last** point in the tab order
* You can also provide a default value: `${1:100}`

**Examples from the docs:**
```sql
-- Simple snippet (no insertion points)
WHERE fare_amount > 100

-- Snippet with an insertion point
WHERE fare_amount > ${1:value}

-- Snippet with a default value
WHERE fare_amount > ${1:100}

-- Snippet with multiple insertion points
WHERE fare_amount > ${2:min_value} AND fare_amount < ${1:max_value} AND trip_distance < ${0:max_distance}
```

---

**Tip:** Use descriptive names in the **Replace** field that are easy to remember

   
### Step 16: Use Snippets in Your Queries

**Try this:**

Query snippets can be used in the **SQL editor**, **notebook SQL cells**, **SQL files**, and **AI/BI dashboard queries**.

1. Open a new query

2. Start typing the snippet name: `datefilter`

3. Select it from the **autocomplete** window
   * In the SQL editor, you can also manually open autocomplete with **Control+Space** and select a snippet
   * You may need to toggle off "New SQL editor" as there are some quirks still being worked out

4. The snippet code appears:
```sql
WHERE DATE(date_column) >= 'start_date'
  AND DATE(date_column) <= 'end_date'
```

5. **Insertion points** are highlighted:
   * The text insertion caret jumps to the **first insertion point**
   * Type the actual column name
   * Press **Tab** to jump to the next insertion point
   * Type the start_date value
   * Press **Tab** again to jump to the next insertion point
   * Type the end_date value
   * An insertion point of `${0}` is always the **last** point in the tab order

6. Complete query:
```sql
SELECT *
FROM samples.nyctaxi.trips
WHERE DATE(tpep_pickup_datetime) >= '2016-01-01'
  AND DATE(tpep_pickup_datetime) <= '2016-01-31'
LIMIT 10
```
---

**Tip:** Snippets save time on repetitive SQL patterns and work across all SQL editing surfaces

## Troubleshooting Advanced Features

**Parameters not working:**
* Check syntax: `:parameter_name` (colon prefix)
* Verify parameter is configured
* Check for typos in parameter name
* Named parameter markers handle typing automatically (no quotes needed around date/string params)

**Visualizations not showing:**
* Verify query returns data
* Check column names match visualization config
* Ensure data types are appropriate for chart type
* Try recreating the visualization

**Schedule not running:**
* Verify schedule is enabled (not paused)
* Check warehouse is available
* Verify query is saved
* Check execution history for errors
* Ensure timezone is correct

**Alerts not triggering:**
* Verify the alert schedule is active
* Check the alert query runs successfully (use **Run now**)
* Use **Test condition** to validate the threshold
* Check that the condition, aggregation, and column are set correctly
* Ensure subscribers are added in the Notifications section
* Check spam folder for notification emails
* Review **Alert history** for past evaluation results

**Snippets not appearing:**
* Verify snippet is saved
* Check trigger word spelling
* Try pressing `Ctrl+Space` / `Cmd+Space`
* Refresh the page
* Toggle off "New SQL editor"

**Multiple statements issues:**
* Ensure semicolons separate statements
* Check each statement is valid SQL
* Verify no syntax errors
* Test statements individually first
